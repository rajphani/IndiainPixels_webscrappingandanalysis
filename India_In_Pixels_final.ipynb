```
My_players = soup.findAll('div',{'class':'ciPlayerbycapstable'})
Players=[] 
players = My_players[0].findAll('ul')
players = players[0].findAll('li')
for player in players:
  if(len(player.findAll('li'))>2):
    Players.append(player.findAll('li')[1].text)
```

```

def cum_sum(player_id, year=2019):
#for player_id in player_ids:
   
  URRL="http://stats.espncricinfo.com/ci/engine/player/playerid.html?class=2;template=results;type=batting;view=match"
  URRL=URRL.replace("playerid",player_id)
  website_url = requests.get(URRL).text
  soup=[]
  soup = BeautifulSoup(website_url,"lxml")
  My_runs = []
  My_runs = soup.findAll('table',{'class':'engineTable'})
  
  cols=[]
  years=pd.Series()
  runs=pd.Series()
  Player_run_year_df=pd.DataFrame()
  table_body = []
  table_body=My_runs[3].find('tbody')
   
  rows=[]
  rows = table_body.find_all('tr')

  for row in rows: 
    #if(row.find_all('td')[7].text.__contains__('XI')==False):
    cols.append([row.find_all('td')[1].text,row.find_all('td')[9].text[-4:]])
  
  
  years=pd.Series([x[1] for x in cols]).replace(to_replace="-",value=0)
  runs=pd.Series([x[0] for x in cols]).replace(to_replace="-",value=0)
  
  years=pd.to_numeric(years)
  runs=pd.to_numeric(runs)
  frame = {  'Year': years ,'Runs': runs} 

  Player_run_year_df=pd.DataFrame()
  Player_run_year_cumm=pd.DataFrame()
  Player_run_year_df=pd.DataFrame(frame)
  cond = (Player_run_year_df['Year']<=int(year))
  Player_run_year_cumm=Player_run_year_df.where(cond)
  
  print( Player_run_year_cumm["Runs"].sum())
```
